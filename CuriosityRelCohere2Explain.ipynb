🧠 Curiosity-Driven Knowledge Exploration System
Overview
This project is a curiosity-driven knowledge exploration system named Curiosity, inspired by the concepts outlined in the paper "Curiosity Is All You Need: A Comprehensive Framework for Intrinsically Motivated Language Models".
The system autonomously explores concepts and relationships, building a dynamic knowledge graph. It uses a combination of probabilistic reasoning, reinforcement learning, and natural language processing to generate hypotheses, refine its understanding, and produce a reflective narrative of its intellectual journey. The output is presented in an HTML report, which includes a visual representation of the knowledge graph, a list of hypotheses, and a creative essay.
The system is designed to be highly observable, addressing the opacity of traditional Large Language Models (LLMs) by providing transparency into its reasoning process. Each hypothesis is annotated with confidence scores, frequency metrics, and relationship classifications, making the system's outputs interpretable and trustworthy.
Features
🔍 Curiosity-Driven Exploration

The system autonomously explores concepts and relationships, driven by curiosity scores that balance novelty, uncertainty, and reward
It generates hypotheses about relationships between concepts and validates them using an LLM

🕸️ Dynamic Knowledge Graph

The system builds and updates a knowledge graph that visually represents the relationships between concepts
Each node (concept) and edge (relationship) is annotated with metadata such as confidence scores, frequency, and classification types

🧪 Hypothesis Tracking

The system tracks and numbers each hypothesis, providing a clear audit trail of its reasoning process
Hypotheses are classified into types (e.g., Supertype-Subtype, Causal, Functional, etc.), adding semantic context to the knowledge graph

📝 Reflective Narrative

The system generates a creative essay that reflects on its intellectual journey, tying together hypotheses and insights
This narrative provides a human-readable summary of the system's exploration

📊 HTML Report
The system produces an HTML report that includes:

A visual representation of the knowledge graph
A list of seed concepts and their relationships
A detailed summary of hypotheses, including confidence scores and classifications
A creative essay that narrates the system's exploration

👁️ Observability

The system is highly observable, providing transparency into its reasoning process
This makes the system's outputs interpretable and trustworthy, addressing the opacity of traditional LLMs

Prototype Nature and Constraints
Prototype Showcasing a Concept
This system is a prototype designed to showcase the power of curiosity-driven learning and dynamic knowledge graph construction. It demonstrates how AI can autonomously explore and reason about abstract concepts, generate hypotheses, and reflect on its own learning process.
Artificial Constraint: 45 Hypotheses
To make the prototype manageable and focused, the system is artificially constrained to generate a maximum of 45 hypotheses. This constraint ensures that the exploration process is concise and that the output can be easily analyzed and interpreted.
Infinite Learning Potential
Unlike traditional LLMs, which are constrained to a fixed dataset and a specific point in time, this system has the potential to learn infinitely and in a greedy manner. It can continuously update its knowledge graph in real time, incorporating new information and refining its understanding as it explores.
Real-Time Learning
The system is designed to learn in real time, updating its knowledge graph on the fly as new hypotheses are generated and validated. This makes it a powerful tool for dynamic and evolving domains, where new information is constantly being discovered.

```python
# =========================
# Physics-Inspired Coherent AI System
# =========================
!pip install ratelimit networkx torch sentence-transformers matplotlib numpy scipy google-generativeai
import math
import re
import random
import re
import time
import hashlib
import networkx as nx
import torch
import numpy as np
import matplotlib.pyplot as plt
from sentence_transformers import SentenceTransformer
from ratelimit import limits, sleep_and_retry
from google.colab import userdata
import google.generativeai as genai
from IPython.display import Image, HTML
import datetime
import io
import base64
from scipy.linalg import eig
from collections import defaultdict
from urllib.parse import quote
import html

# ========================
# Configuration
# ========================
BLUE = '\033[94m'
RED = '\033[91m'
RESET = '\033[0m'
ANSI_TO_HTML = {
    BLUE: 'blue',
    RED: 'red',
    RESET: 'black'
}
html_output_lines = []

# ========================
# Core Classes
# ========================

class CoherenceEngine:
    """Implements laser-inspired coherence mechanics"""
    def __init__(self, intelligence):
        self.intel = intelligence
        self.K = 0.15  # Coupling strength
        self.α = 0.85  # Confidence decay
        self.θ = 0.65  # Coherence threshold
        self.γ = 0.6   # Centrality threshold
        self.coherence_history = []

    def initialize_phases(self):
        """Initialize phase and amplitude states"""
        for node in self.intel.knowledge_graph.nodes:
            self.intel.node_phases[node] = random.uniform(0, 2 * math.pi)
            self.intel.node_amplitudes[node] = self.intel.rewards.get(node, 0.5)

    def update_dynamics(self, steps=3):
        """Execute full coherence update cycle"""
        self.update_phases(steps)
        self.update_amplitudes()
        self.record_coherence()

    def update_phases(self, steps=3):
        """Kuramoto-like phase synchronization"""
        for _ in range(steps):
            new_phases = {}
            for node in self.intel.knowledge_graph.nodes:
                ω = 0.05 * (self.intel.rewards.get(node, 0.5) - 0.5)
                coupling = 0
                neighbors = list(self.intel.knowledge_graph.neighbors(node))

                for neighbor in neighbors:
                    if (node, neighbor) in self.intel.knowledge_graph.edges:
                        c_ij = self.intel.knowledge_graph.edges[node, neighbor].get('confidence', 0)
                        phase_diff = self.intel.node_phases[neighbor] - self.intel.node_phases[node]
                        coupling += c_ij * math.sin(phase_diff)

                coupling *= self.K / max(1, len(neighbors))
                new_phases[node] = self.intel.node_phases[node] + ω + coupling

            for node, phase in new_phases.items():
                self.intel.node_phases[node] = phase % (2 * math.pi)

    def update_amplitudes(self):
        """Update amplitudes via eigenvector centrality"""
        nodes = list(self.intel.knowledge_graph.nodes)
        n = len(nodes)
        if n == 0: return

        conf_matrix = np.zeros((n, n))
        for i, u in enumerate(nodes):
            for j, v in enumerate(nodes):
                if self.intel.knowledge_graph.has_edge(u, v):
                    conf_matrix[i,j] = self.intel.knowledge_graph.edges[u,v].get('confidence', 0)

        np.fill_diagonal(conf_matrix, 0.1)
        conf_matrix = conf_matrix / np.max(conf_matrix)

        try:
            vals, vecs = eig(conf_matrix)
            max_idx = np.argmax(np.real(vals))
            centrality = np.abs(np.real(vecs[:, max_idx]))
            centrality = centrality / np.max(centrality)

            for i, node in enumerate(nodes):
                self.intel.node_amplitudes[node] = (
                    0.7 * self.intel.node_amplitudes.get(node, 0.5) +
                    0.3 * centrality[i]
                )
        except Exception as e:
            print(f"Eigenvector calculation failed: {str(e)}")

    def calculate_coherence(self):
        """Compute current coherence metric C"""
        real_sum = imag_sum = 0
        total_amplitude = 0
        for node in self.intel.knowledge_graph.nodes:
            A = self.intel.node_amplitudes.get(node, 0)
            ϕ = self.intel.node_phases.get(node, 0)
            real_sum += A * math.cos(ϕ)
            imag_sum += A * math.sin(ϕ)
            total_amplitude += A

        N = len(self.intel.knowledge_graph.nodes)
        return math.sqrt(real_sum**2 + imag_sum**2) / total_amplitude if N > 0 else 0

    def record_coherence(self):
        """Track coherence over time"""
        self.coherence_history.append(self.calculate_coherence())

    def generate_coherent_beam(self):
        """Form coherent beam of aligned concepts"""
        beam = []
        real_sum = imag_sum = 0

        sorted_nodes = sorted(
            self.intel.knowledge_graph.nodes,
            key=lambda x: -self.intel.node_amplitudes.get(x, 0)
        )

        for node in sorted_nodes:
            A = self.intel.node_amplitudes.get(node, 0)
            ϕ = self.intel.node_phases.get(node, 0)

            proj_real = real_sum + A*math.cos(ϕ)
            proj_imag = imag_sum + A*math.sin(ϕ)
            proj_magnitude = math.sqrt(proj_real**2 + proj_imag**2)

            if proj_magnitude >= max(self.θ, math.sqrt(real_sum**2 + imag_sum**2)):
                beam.append(node)
                real_sum = proj_real
                imag_sum = proj_imag

        return beam

class CuriosityModule:
    """Implements curiosity-driven exploration"""
    def __init__(self, intelligence):
        self.intel = intelligence
        self.learning_priority = {}
        self.hypothesis_counter = 0
        self.coherence_engine = CoherenceEngine(intelligence)

    def calculate_curiosity(self, concept):
        """Composite curiosity score with coherence awareness"""
        try:
            uncertainty = 1 - self._get_confidence(concept)
            reward_grad = abs(self.intel.rewards.get(concept, 0.5) - 0.5)
            novelty = self._calculate_novelty(concept)
            phase_alignment = math.cos(self.intel.node_phases.get(concept, 0))
            centrality = self.intel.node_amplitudes.get(concept, 0)

            return (
                0.25 * uncertainty +
                0.2 * reward_grad +
                0.3 * novelty +
                0.15 * phase_alignment +
                0.1 * centrality
            )
        except Exception as e:
            print(f"Curiosity calculation error: {str(e)}")
            return 0.5

    def _get_confidence(self, concept):
        response = self.intel.probabilistic_reasoning(
            concept,
            "your internal understanding and the information in the knowledge graph"
        )
        return max(0.0, min(1.0, response))

    def _calculate_novelty(self, concept):
        try:
            if concept not in self.intel.embeddings: return 1.0
            base_embed = self.intel.embeddings[concept]
            similarities = [
                torch.cosine_similarity(base_embed, embed, dim=0).item()
                for other, embed in self.intel.embeddings.items()
                if other != concept
            ]
            return 1 - max(similarities) if similarities else 1.0
        except:
            return 0.5

    def explore(self, top_n=2):
        """Execute curiosity-driven exploration"""
        try:
            self.learning_priority = {
                c: self.calculate_curiosity(c)
                for c in self.intel.knowledge_graph.nodes
            }

            targets = sorted(
                self.learning_priority.items(),
                key=lambda x: x[1],
                reverse=True
            )[:top_n]

            for concept, score in targets:
                print(f"Exploring {concept} (curiosity={score:.2f})")
                self.targeted_knowledge_expansion(concept)

            return True
        except Exception as e:
            print(f"Exploration failed: {str(e)}")
            return False

    def targeted_knowledge_expansion(self, base_concept):
        """Expand knowledge from a base concept"""
        prompts = [
            f"Suggest 3 specific concepts related to '{base_concept}' using EXACT format:\n"
            "Concept: [Name], Relation: [Relationship]\n"
            "Example:\n"
            "Concept: Black Holes, Relation: extreme manifestation of\n"
            "Only return concepts in this format, no other text!",

            f"Generate novel associations for '{base_concept}' using format:\n"
            "Concept: [Name], Relation: [Relationship]\n"
            "Important: Use only the exact format with no additional explanations!",

            f"Propose connected concepts for '{base_concept}' using:\n"
            "Concept: [Name], Relation: [Relationship]\n"
            "Return only concept-relation pairs, no other text!"
        ]

        prompt = random.choice(prompts)
        response = self.intel.query_llm(prompt)
        if not response: return False

        parsed = re.findall(
            r'(?i)Concept:\s*([^\n\r]+?)\s*,\s*Relation:\s*([^\n\r]+?)(?=\n|Concept:|$)',
            response
        )

        added_count = 0
        for concept, relation in parsed[:3]:
            clean_concept = self.clean_concept(concept)
            clean_relation = self.clean_relation(relation)

            if (len(clean_concept) >= 3 and
                    len(clean_relation) >= 3 and
                    self.intel.validate_relation(clean_relation)):
                if self.intel.add_concept(clean_concept, {clean_relation: base_concept}):
                    self._update_learning_rewards(base_concept, clean_concept)
                    added_count += 1

        return added_count > 0

    def explore_internal_links(self, top_n=5):
        """Discover relationships between existing concepts"""
        nodes = list(self.intel.knowledge_graph.nodes)
        if len(nodes) < 2: return False

        sorted_curiosity = sorted(
            self.learning_priority.items(),
            key=lambda item: item[1],
            reverse=True
        )
        focused_nodes = [item[0] for item in sorted_curiosity[:top_n]]

        explored_pairs = set()
        added_link = False

        for node_a_focus in focused_nodes:
            for node_b in nodes:
                if node_a_focus == node_b: continue

                pair = tuple(sorted((node_a_focus, node_b)))
                if pair in explored_pairs: continue
                explored_pairs.add(pair)

                prompt_relation = f"""What is a possible relationship between '{node_a_focus}' and '{node_b}'? The relationship should be strictly directional from '{node_a_focus}' to '{node_b}'. Answer by providing a concise phrase describing the relationship from '{node_a_focus}' to '{node_b}', starting with a verb (e.g., 'is a type of', 'causes', 'is related to'). Make sure you don't repeat neither '{node_a_focus}' nor '{node_b}' in your description. IMPORTANT: If there is no direct or significant relationship, state 'No significant relationship'. Never provide multi-sentence answers. If you have many possibilities for answering only state the answer that feels more correct to you, but don't suggest multiple options. Do not finalize your answer with a period. NEVER include the word IT in your answer."""
                response_relation = self.intel.query_llm(prompt_relation)

                if response_relation and "No significant relationship" not in response_relation:
                    relation_phrase = self.clean_relation(response_relation.strip())

                    if self.intel.validate_relation(relation_phrase):
                        prompt_confidence = f"On a scale of 0 to 1, how certain are you about the stated fact '{node_a_focus}' '{relation_phrase}' '{node_b}' being true? Please provide only a numerical value between 0 and 1 with no further explanation."
                        response_confidence = self.intel.query_llm(prompt_confidence)

                        prompt_classification = f"""Classify the relationship '{node_a_focus}' '{relation_phrase}' '{node_b}' using one of the following types: Supertype-Subtype (Is-a), Part-Whole (Has-a / Part-of), Containment (Includes / Belongs-to), Causal (Cause-Effect), Functional (Enables / Prevents), Dependency (Needs / Requires), Usage (Used-for), Influence (Affects / Shapes), Synonymy (Same-as / Equivalent-to), Antonymy (Opposite-of), Analogy (Similar-to / Like), Degree (More/Less than), Sequence (Before-After), Simultaneity (Occurs-with / Coexists-with), Location (Located-in / Found-at), Direction (Moves-toward / Away-from), Equivalence (≡ / If-and-only-if), Implication (→ / If-then), Exclusion (Mutually Exclusive), Set Membership (Element-of / Subset-of), Conceptual Composition (Made-of / Constructed-from), Abstraction (Abstracts-from / Represents), Perspective (Viewed-as), or UNDEFINED. Provide only the type, no additional text."""
                        response_classification = self.intel.query_llm(prompt_classification)

                        confidence = None
                        if response_confidence:
                            numbers = re.findall(r'\d?\.?\d+', response_confidence)
                            if numbers:
                                try:
                                    confidence = float(numbers[0])
                                    confidence = max(0.0, min(1.0, confidence))
                                except ValueError:
                                    print(f"Could not parse confidence score: {response_confidence}")

                        edge_exists = False
                        existing_edge_data = None
                        for u, v, data in self.intel.knowledge_graph.edges(data=True):
                            if u == node_a_focus and v == node_b and data.get('relation') == relation_phrase:
                                edge_exists = True
                                existing_edge_data = data
                                break

                        if edge_exists:
                            if 'frequency' in existing_edge_data:
                                existing_edge_data['frequency'] += 1
                            else:
                                existing_edge_data['frequency'] = 2
                            print(f"Frequency of link {node_a_focus} → {relation_phrase} → {node_b} increased to {existing_edge_data['frequency']}")
                            added_link = True
                        else:
                            self.hypothesis_counter += 1
                            edge_data = {
                                'relation': relation_phrase,
                                'hypothesis': f"Relationship: {response_relation.strip()}",
                                'frequency': 1,
                                'classification': response_classification.strip() if response_classification else 'UNDEFINED'
                            }
                            if confidence is not None:
                                edge_data['confidence'] = confidence
                                edge_data['hypothesis'] += f", Confidence: {response_confidence.strip()}"
                            self.intel.knowledge_graph.add_edge(
                                node_a_focus,
                                node_b,
                                **edge_data
                            )
                            print(f"Hypothesis created: {self.hypothesis_counter}")
                            print(f"Added internal link: {node_a_focus} → {relation_phrase} → {node_b} (Confidence: {confidence if confidence is not None else 'N/A'}, Classification: {edge_data['classification']})")
                            added_link = True
                            self._update_learning_rewards(node_a_focus, node_b)
                            self._update_learning_rewards(node_b, node_a_focus)

                            if confidence is not None and confidence >= 0.8:
                                self.explore_relationship_names(node_a_focus, relation_phrase, node_b)

                            if self.hypothesis_counter >= 10:
                                return added_link
        return added_link

    def explore_relationship_names(self, node_a, relation, node_b):
        """Explore specific names for relationships"""
        prompt = f"""Provide a single, specific name for the relationship '{node_a} {relation} {node_b}'. The name should be a concrete example or instance of this relationship. Do not provide synonyms or alternatives. Respond with only the name, no additional text."""
        response = self.intel.query_llm(prompt)

        if response:
            new_concept = self.clean_concept(response.strip())
            if self.intel.add_concept(new_concept, {"is a type of": relation}):
                print(f"Added new concept: {new_concept} (is a type of {relation})")
                self._update_learning_rewards(node_a, new_concept)
                self._update_learning_rewards(new_concept, node_b)

    def clean_concept(self, text):
        return re.sub(r'[\*\"\']', '', text).strip().title()

    def clean_relation(self, text):
        return re.sub(r'[\*\"\'\(\)]', '', text).strip().lower()

    def _update_learning_rewards(self, base_concept, new_concept):
        self.intel.reinforcement_learning(base_concept, min(1.0, self.intel.rewards[base_concept] + 0.15))
        self.intel.reinforcement_learning(new_concept, 0.7)

class UniversalIntelligence:
    """Main AI system class"""
    def __init__(self):
        self.knowledge_graph = nx.DiGraph()
        self.embeddings = {}
        self.rewards = {}
        self.node_phases = {}
        self.node_amplitudes = {}
        self.embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
        self.generation_config = {
            "temperature": 0.9,
            "top_p": 0.95,
            "top_k": 50,
            "max_output_tokens": 1024
        }
        self.description_cache = {}
        self.learning_interval = 15
        self.last_learned = time.time()
        self.curiosity = CuriosityModule(self)
        self.explored_relations = set()
        self.model_name = 'gemini-1.5-pro'
        self.initial_concepts_data = {}
        self.seed_start_time = 0
        self.seed_end_time = 0
        self.llm_wait_time = 0
        self.rate_limit_wait_time = 0

    def get_cache_key(self, concept):
        return hashlib.md5(concept.encode()).hexdigest()

    def seed_initial_concepts(self, initial_concepts):
        """Initialize the knowledge graph with seed concepts"""
        print("--- Seeding Initial Concepts ---")
        self.seed_start_time = time.time()
        for concept, relations in initial_concepts:
            self.initial_concepts_data[concept] = relations
            if self.add_concept(concept, relations):
                print(f"✓ {concept}")
            else:
                print(f"✗ {concept}")
        self.seed_end_time = time.time()
        self.curiosity.coherence_engine.initialize_phases()

    @sleep_and_retry
    @limits(calls=15, period=60)
    def query_llm(self, prompt):
        """Query the language model with rate limiting"""
        try:
            model = genai.GenerativeModel(self.model_name,
                                        generation_config=self.generation_config)
            response = model.generate_content(prompt)
            return response.text if response.text else None
        except Exception as e:
            if "rate limit" in str(e).lower() or "quota" in str(e).lower():
                print(f"Rate limit reached. Error: {str(e)}")
                time.sleep(60)
                raise
            else:
                print(f"API error: {str(e)}")
                return None

    def add_concept(self, concept, relations=None):
        """Add a new concept to the knowledge graph"""
        try:
            concept = concept.strip().title()
            if not re.match(r"^[\w\s'\-\(\)]{3,50}$", concept):
                print(f"Invalid concept format: {concept}")
                return False

            cache_key = self.get_cache_key(concept)
            if cache_key not in self.description_cache:
                description = self.query_llm(f"Describe the concept of {concept} in one sentence. Make sure you're describing a concept. Don't provide anything else.")
                if not description: return False
                self.description_cache[cache_key] = description

            embedding = self.embedding_model.encode(self.description_cache[cache_key])
            self.embeddings[concept] = torch.tensor(embedding)
            self.rewards[concept] = 0.5
            self.node_phases[concept] = random.uniform(0, 2 * math.pi)
            self.node_amplitudes[concept] = 0.5
            self.knowledge_graph.add_node(concept)
            print(f"Added concept: {concept}")

            if relations:
                for rel, target in relations.items():
                    target = target.strip().title()
                    if not self.validate_relation(rel) or not target:
                        continue

                    if target not in self.embeddings:
                        self.add_concept(target)

                    if nx.has_path(self.knowledge_graph, target, concept):
                        print(f"Prevented circular relation: {target} → {concept}")
                        continue

                    relation_tuple = (concept, rel, target)
                    if relation_tuple not in self.explored_relations:
                        self.knowledge_graph.add_edge(concept, target, relation=rel, confidence=0.7, frequency=1)
                        self.explored_relations.add(relation_tuple)
                        print(f"Added relation: {concept} → {rel} → {target}")
            return True
        except Exception as e:
            print(f"Error adding {concept}: {str(e)}")
            return False

    def validate_relation(self, relation):
        """Validate relationship phrases"""
        relation = relation.lower().strip()
        banned_terms = {'associated', 'linked', 'connection'}
        return (
            re.match(r"^[\w\s-]{3,30}$", relation) and
            not any(term in relation for term in banned_terms) and
            len(relation.split()) <= 4
        )

    def probabilistic_reasoning(self, concept, reasoning_context):
        """Query LLM for confidence estimate"""
        prompt = f"On a scale of 0 to 1, based on '{reasoning_context}', how certain are you about the concept of '{concept}'? Please provide only the numerical value."
        response = self.query_llm(prompt)
        numbers = re.findall(r'\d?\.?\d+', response or "")
        return max(0.0, min(1.0, float(numbers[0]))) if numbers else 0.5

    def reinforcement_learning(self, concept, reward):
        """Update concept rewards"""
        if concept in self.rewards:
            self.rewards[concept] = 0.85 * self.rewards[concept] + 0.15 * reward
            self.rewards[concept] = max(0.1, min(0.9, self.rewards[concept]))

    def autonomous_learning_cycle(self):
        """Execute full learning cycle"""
        if time.time() - self.last_learned >= self.learning_interval:
            print("\n=== Learning Cycle ===")
            if self.curiosity.explore(top_n=3):
                pass
            if self.curiosity.explore_internal_links(top_n=5):
                pass

            self.curiosity.coherence_engine.update_dynamics()
            current_coherence = self.curiosity.coherence_engine.calculate_coherence()
            print(f"Current coherence: {current_coherence:.2f}")

            if current_coherence > 0.6:
                beam = self.curiosity.coherence_engine.generate_coherent_beam()
                print(f"Coherent beam formed: {beam}")

            self.show_graph_stats()
            self.last_learned = time.time()

    def show_graph_stats(self):
        """Display knowledge graph statistics"""
        print("\nKnowledge Graph Summary:")
        print(f"Nodes: {len(self.knowledge_graph.nodes)}")
        print(f"Edges: {len(self.knowledge_graph.edges)}")
        print("Top Concepts by Reward:")
        for concept in sorted(self.rewards, key=self.rewards.get, reverse=True)[:5]:
            print(f"- {concept}: {self.rewards[concept]:.2f}")

# ========================
# Visualization & Reporting
# ========================

def visualize_knowledge_graph(intelligence):
    """Visualize the knowledge graph with physics-inspired layout"""
    plt.figure(figsize=(12, 12))

    # Create axis explicitly
    ax = plt.gca()

    # Position nodes using phase-aware layout
    pos = {
        node: (
            intelligence.node_amplitudes[node] * math.cos(intelligence.node_phases[node]),
            intelligence.node_amplitudes[node] * math.sin(intelligence.node_phases[node])
        )
        for node in intelligence.knowledge_graph.nodes
    }

    # Draw nodes with size proportional to amplitude
    node_sizes = [
        3000 * (intelligence.node_amplitudes.get(node, 0.5) ** 2)
        for node in intelligence.knowledge_graph.nodes
    ]

    # Create node collection for color mapping
    nodes = nx.draw_networkx_nodes(
        intelligence.knowledge_graph,
        pos,
        node_size=node_sizes,
        node_color=list(intelligence.rewards.values()),
        cmap=plt.cm.plasma,
        alpha=0.8,
        ax=ax
    )

    # Draw edges with width proportional to confidence
    edge_widths = [
        3 * data.get('confidence', 0.5)
        for _, _, data in intelligence.knowledge_graph.edges(data=True)
    ]

    nx.draw_networkx_edges(
        intelligence.knowledge_graph,
        pos,
        width=edge_widths,
        alpha=0.5,
        edge_color='gray',
        ax=ax
    )

    # Label nodes
    nx.draw_networkx_labels(
        intelligence.knowledge_graph,
        pos,
        font_size=10,
        font_weight='bold',
        ax=ax
    )

    # Add phase indicators
    for node, (x, y) in pos.items():
        phase = intelligence.node_phases[node]
        ax.plot(
            [x, x + 0.1 * math.cos(phase)],
            [y, y + 0.1 * math.sin(phase)],
            'r-',
            linewidth=2
        )

    plt.title("Knowledge Graph with Phase Alignment")

    # Add colorbar using the node collection
    plt.colorbar(nodes, label='Reward', ax=ax)
    plt.tight_layout()
    plt.show()

def generate_beam_essay(intelligence, beam):
    """Generate narrative about the coherent beam"""
    if not beam: return "No coherent beam formed"

    beam_description = "\n".join([
        f"- {node} (A: {intelligence.node_amplitudes.get(node, 0):.2f}, ϕ: {intelligence.node_phases.get(node, 0):.2f})"
        for node in beam
    ])

    prompt = f"""The following concepts have formed a coherent beam through phase alignment. Write a compelling essay explaining how these concepts relate to each other and what insights emerge from their alignment. Focus on the most significant relationships.

Beam Components:
{beam_description}

Essay:"""
    return intelligence.query_llm(prompt) or "Failed to generate essay"

def generate_html_report(intelligence):
    """Generate comprehensive HTML report with all requested sections"""
    try:
        # Generate core components
        beam = intelligence.curiosity.coherence_engine.generate_coherent_beam() or []
        beam_essay = generate_beam_essay(intelligence, beam) or "No insights generated"

        # URL for the coherence image
        coherence_img_url = "https://media.licdn.com/dms/image/v2/D4E22AQG785Lvveetkg/feedshare-shrink_800/B4EZXKlRPaG0Ao-/0/1742860522790?e=1745452800&v=beta&t=bdcsR6dRV5rlVynNWs0n5af7o2j_WjbyHoT5DZqmvOk"

        # --- Visualization Section ---
        fig = plt.figure(figsize=(16, 12))
        ax = fig.add_subplot(111)

        # Generate positions using phase/amplitude
        pos = {
            node: (
                intelligence.node_amplitudes[node] * math.cos(intelligence.node_phases[node]),
                intelligence.node_amplitudes[node] * math.sin(intelligence.node_phases[node])
            )
            for node in intelligence.knowledge_graph.nodes
        }

        # Draw nodes with reward coloring
        node_sizes = [
            3000 * (intelligence.node_amplitudes.get(node, 0.5) ** 2)
            for node in intelligence.knowledge_graph.nodes
        ]
        node_colors = [
            intelligence.rewards.get(node, 0.5)
            for node in intelligence.knowledge_graph.nodes
        ]

        nodes = nx.draw_networkx_nodes(
            intelligence.knowledge_graph,
            pos,
            node_size=node_sizes,
            node_color=node_colors,
            cmap=plt.cm.plasma,
            alpha=0.8,
            ax=ax
        )

        # Draw edges
        nx.draw_networkx_edges(
            intelligence.knowledge_graph,
            pos,
            width=[3 * data.get('confidence', 0.5)
                 for _, _, data in intelligence.knowledge_graph.edges(data=True)],
            alpha=0.5,
            edge_color='gray',
            ax=ax
        )

        # Draw labels
        nx.draw_networkx_labels(
            intelligence.knowledge_graph,
            pos,
            font_size=10,
            font_weight='bold',
            ax=ax
        )

        # Add phase direction indicators
        for node in intelligence.knowledge_graph.nodes:
            x, y = pos[node]
            phase = intelligence.node_phases[node]
            ax.plot(
                [x, x + 0.1 * math.cos(phase)],
                [y, y + 0.1 * math.sin(phase)],
                'r-',
                linewidth=2,
                alpha=0.7
            )

        # Add reward colorbar
        cbar = plt.colorbar(nodes, ax=ax, shrink=0.8)
        cbar.set_label('Concept Reward Value', rotation=270, labelpad=15)

        # Save to buffer
        img_buffer = io.BytesIO()
        plt.savefig(img_buffer, format='png', bbox_inches='tight', dpi=120)
        plt.close()
        img_base64 = base64.b64encode(img_buffer.getvalue()).decode('utf-8')

        # --- Prepare Hypotheses Data ---
        hypotheses = []
        for u, v, data in intelligence.knowledge_graph.edges(data=True):
            if 'hypothesis' in data:
                hypotheses.append(f"""
                <div class="hypothesis">
                    <b>Connection:</b> {html.escape(f"{u} → {v}")}<br>
                    <b>Type:</b> {html.escape(data.get('classification', 'N/A'))}<br>
                    <b>Confidence:</b> {html.escape(f"{data.get('confidence', 0.5):.2f}")}<br>
                    <b>Details:</b> {html.escape(data['hypothesis'])}
                </div>
                """)

        # --- Construct HTML ---
        timestamp = datetime.datetime.now().strftime("%d%m%Y%H%M%S")
        filename = f"{timestamp}.html"

        # Build HTML parts safely
        html_parts = []
        html_parts.append("""<html>
<head>
    <title>Coherent AI Report</title>
    <style>
        body { font-family: Arial, sans-serif; line-height: 1.6; padding: 20px; }
        .section { margin-bottom: 30px; padding-bottom: 20px; border-bottom: 1px solid #eee; }
        .graph-img { max-width: 100%; height: auto; border: 1px solid #ddd; }
        .hypothesis {
            background: #f8f8f8;
            padding: 10px;
            margin: 10px 0;
            border-radius: 5px;
            border-left: 4px solid #4285F4;
        }
        .explanation { color: #666; font-style: italic; }
        .image-container { text-align: center; margin: 20px 0; }
    </style>
</head>
<body>
    <h1>Knowledge Evolution Report</h1>
    <p class="explanation">Generated: """ +
    html.escape(datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")) +
    """</p>

    <!-- Coherence Image Section -->
    <div class="section">
        <h2>The Principle of Coherence</h2>
        <div class="image-container">
            <img src=\"""" + html.escape(coherence_img_url) + """\"
                 alt=\"""" + html.escape("Coherence Principle") + """\"
                 style=\"max-width: 100%; height: auto; border: 1px solid #ddd; border-radius: 4px;\">
            <p class=\"explanation\">""" +
            html.escape("Visual representation of the coherence principle that inspires this system's architecture") +
            """</p>
        </div>
    </div>

    <!-- 1. Initial Seeds -->
    <div class="section">
        <h2>Initial Knowledge Seeds</h2>
        <ul>""")

        # Add initial seeds with proper escaping
        for k, relations in intelligence.initial_concepts_data.items():
            connections = []
            for rel, v in relations.items():
                connections.append(html.escape(str(v)) + " (" + html.escape(rel) + ")")
            html_parts.append("<li><b>" + html.escape(str(k)) + "</b> connected to: " + ', '.join(connections) + "</li>")

        html_parts.append("""
        </ul>
        <p class=\"explanation\">These were the foundational concepts provided to the system.</p>
    </div>

    <!-- 2. Discovered Hypotheses -->
    <div class="section">
        <h2>Discovered Relationships</h2>""")

        # Add hypotheses
        if hypotheses:
            html_parts.extend(hypotheses)
        else:
            html_parts.append("""<p>No new hypotheses were confirmed.</p>""")

        html_parts.append("""
        <p class=\"explanation\">Relationships established through exploration with confidence > 0.7</p>
    </div>

    <!-- 3. Knowledge Graph -->
    <div class="section">
        <h2>Phase-Aligned Knowledge Graph</h2>
        <img class="graph-img" src="data:image/png;base64,""" +
        html.escape(img_base64) +
        """\">

        <h3>How to Interpret This Visualization</h3>
        <ul>
            <li><b>Node Size</b>: Concept amplitude (system-assigned importance)</li>
            <li><b>Node Color</b>: Reward value (darker = higher reinforcement)</li>
            <li><b>Red Lines</b>: Phase direction vectors (alignment between concepts)</li>
            <li><b>Edge Thickness</b>: Relationship confidence</li>
        </ul>
        <p class=\"explanation\">
            Clusters of aligned vectors indicate coherent concept groupings that may form beams.
            The colorbar shows the reward scale from least (0.1) to most (0.9) reinforced.
        </p>
    </div>

    <!-- 4. Emergent Insights -->
    <div class="section">
        <h2>Emergent Insights</h2>
        <h3>About This Synthesis</h3>
        <p class=\"explanation\">
            This essay was generated by analyzing the coherent beam - concepts that achieved
            phase alignment (C > 0.65). It represents the system's interpretation of how
            these concepts fundamentally interrelate.
        </p>
        <div style=\"background: #f9f9f9; padding: 15px; border-radius: 5px;\">""" +
        html.escape(beam_essay).replace("\n", "<br>") +
        """
        </div>
    </div>
</body>
</html>""")

        # Combine all parts
        html_content = "".join(html_parts)

        # Save and provide download
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(html_content)

        from IPython.display import display, HTML
        display(HTML(f"""
        <a download="{html.escape(filename)}" href="data:text/html;charset=utf-8,{quote(html_content)}"
           style="display: inline-block; padding: 10px 15px; background: #4285F4; color: white;
                  border-radius: 5px; text-decoration: none;">
           Download Complete Report
        </a>
        """))

        return filename

    except Exception as e:
        print(f"Report generation failed: {str(e)}")
        return None

########### MAIN SECTION

if __name__ == "__main__":
    # Initialize system
    try:
        GENAI_API_KEY = userdata.get('GOOGLE_API_KEY')
        genai.configure(api_key=GENAI_API_KEY)
    except:
        GENAI_API_KEY = None
        print("Running in demo mode without LLM access")

    ui = UniversalIntelligence()

    # Seed initial concepts
    initial_concepts = [
        ("Consciousness", {"involves": "Subjective Experience"}),
        ("Neural Activity", {"correlates with": "Conscious States"}),
        ("Curiosity", {"involves": "Intrinsic Motivation"}),
        ("Absolute certainty", {"is contrary to": "Curiosity"}),
        ("Coherence", {"is a necessary condition for": "Conscious Unity"})
    ]
    ui.seed_initial_concepts(initial_concepts)

    # Run learning cycles
    try:
        for cycle in range(10):
            print(f"\n=== Learning Cycle {cycle + 1} ===")
            ui.autonomous_learning_cycle()
            time.sleep(5)

            if cycle % 3 == 0:
                visualize_knowledge_graph(ui)

    finally:
        # Generate final report
        report_file = generate_html_report(ui)
        print(f"\nReport generated: {report_file}")
```
